## Working on SVM

```python
import numpy as np
import pandas as pd
from sklearn import preprocessing
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn.metrics import accuracy_score,confusion_matrix
```

```python
hr=pd.read_csv('C:/Users/Administrator/Desktop/Machine learning/with karthirmani/datasets/HR Analytics.csv')
hr=pd.get_dummies(hr)
hr.head()
```

```python
hr1=pd.DataFrame(preprocessing.StandardScaler().fit(hr).transform(hr))
hr1
```

```python
y=hr['Attrition']
xtrain,ytrain,xtest,ytest=train_test_split(hr1,y,test_size=0.3,random_state=100)

model=svm.SVC(gamma=0.001,C=100)
model.fit(xtrain,ytrain)
pred=model.predict(xtest)
print(confusion_matrix(ytest,pred,labels=[0,1]))
```

## using images

```python
from sklearn.datasets import fetch_olivetti_faces
import pandas as pd
import numpy as np
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
data = pd.DataFrame(fetch_olivetti_faces()['data'])
print(data.shape)
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
scaler.fit(data)
data_scaled=pd.DataFrame(scaler.transform(data),columns=data.columns)
```

```python
data_scaled.shape
```

```python
plt.imshow(np.array(data.iloc[0]).reshape(64,64),cmap = 'gray')
plt.figure(figsize=(10,12))
for i in range(100):
    image = np.array(data.iloc[i]).reshape(64,64)
    ax = plt.subplot(10,10, i+1)
    ax.imshow(image, cmap = 'gray')
```

```python
evalues,evectors=np.linalg.eig(data.corr())
```

```python
pca=PCA()
pca.fit(data)
df_pcs=pd.DataFrame(pca.components_)
df_pcs.shape
```

```python
plt.figure(figsize=(10,12))
for i in range(100):
    image = np.array(df_pcs.iloc[i]).reshape(64,64)
    ax = plt.subplot(10,10, i+1)
    ax.imshow(image, cmap = 'gray')
```
